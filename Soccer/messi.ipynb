```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import sklearn 
```

```python
df = pd.read_csv("results.csv")
```

```python
df
```

```python
# finding columns and rows with NaN values, will drop 
nan_count = df.isna().sum()

print(nan_count)
```

```python
# dropping rows with NaN
df_dropped = df.dropna()
df_dropped
```

```python
average_goals_home_team = df_dropped["home_score"].mean()
average_goals_away_team = df_dropped["away_score"].mean()

print(f"Average number of goals for the home team: {average_goals_home_team}")
print(f"Average number of goals for the away team: {average_goals_away_team}")

```

```python
sns.regplot(x='home_score', y='away_score', data=df_dropped)
plt.show()
r = df_dropped['home_score'].corr(df_dropped['neutral'])
print(r)
```
